```python
df=df.drop('index',axis=1)
```

```python
# to make a string to a list of string

df['Description']=df['Description'].apply(lambda x:x.split())
```

```python
df['Reason']=df['Reason'].apply(lambda x:x.split())
```

```python
df
```

```python
# to take only the first 5 words from drug name

df['Drug_Name']=df['Drug_Name'].str.split(' ').str.slice(0,4).str.join(' ')
```

```python
df
```

```python
# to create a separate column concating the 'Reason' & 'Description' 

df['new']=df['Reason']+df['Description']
```

```python
df
```

```python
df=df[['Drug_Name','Description','new']]
```

```python
df
```

```python
# now we need to convert the list into a string

df['new']=df['new'].apply(lambda x:" ".join(x))
```

```python
df
```

```python
#to convert the new column to lower

df['new']=df['new'].apply(lambda x:x.lower())
```

```python
df
```

```python
# to stem(removing the prefix & suffix to find word root) the data

import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

df['new']=df['new'].apply(stem)
```

```python
# to Vectorize (I'm using Bag of Words Vectorizer)

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')

vector=cv.fit_transform(df['new']).toarray()
```

```python
vector[1]
```

```python
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vector)
```

```python
similarity
```

##### Now lets Predict similar medicines from description given by user

```python
def find_index(text):  # to find the index of the 
    text=text.lower()
    t_list=text.split(' ')
    c=0
    for i in df['Description']:
        for element in t_list:
            if element not in i:
                c=c+1
                break
            else: 
                return c+1
    return []
```

```python
def recommend(med):
    id=find_index(med)
    dis=similarity[id]
    med_list=sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[1:6]
    print("Recommended Medicines :-")
    c=1
    for i in med_list:
        print('%d. '%(c) +df.iloc[i[0]].Drug_Name)
        c=c+1
```

```python
text="Headache Migraine Fever"
recommend(text)
```

```python
text="Mild to Modarate Acne"
recommend(text)
```

```python
text="Itching"
recommend(text)
```

```python
text="Type 2 Diabetes Mellitus"
recommend(text)
```

```python
text='chickenpox'
recommend(text)
```
